```python
import random
import json
```

```python
numPrisoners = 100
maxGuesses = 50
numPlays = 10000

successes = 0
attempts = 0
if True:
    print("Trying using a random selection strategy:")
    for i in range(0,numPlays):
        remainingNumbers = []
        prisoners = []
        for i in range(0,numPrisoners):
            remainingNumbers.append(i)
            prisoners.append(i)
        boxes = []
        for ii in range(0,numPrisoners):
            l = len(remainingNumbers)
            r = random.randint(0,l-1)
            n = remainingNumbers[r]
            boxes.append(n)
            del remainingNumbers[r]
        ## Assume success unless someone fails to find a number
        success = True
        for p in prisoners:

            ## (Re-)Set up one box index per prisoner:
            remainingBoxes = prisoners.copy()
            numGuesses = 0
            found = False
            while(numGuesses<maxGuesses):
                numGuesses += 1
                n = random.randint(0,len(remainingBoxes)-1)
                g = remainingBoxes[n]
                del remainingBoxes[n]
                if (g==p):
                    found = True
                    break

            if not found:
                success = False
                break
        attempts += 1
        if (success):
            successes += 1
    print("Successes:", successes, "/", attempts)

    
successes = 0
attempts = 0
if True:
    print("Trying using the chaining strategy:")
    for i in range(0,numPlays):
        remainingNumbers = []
        prisoners = []
        for i in range(0,numPrisoners):
            remainingNumbers.append(i)
            prisoners.append(i)
        boxes = []
        for ii in range(0,numPrisoners):
            l = len(remainingNumbers)
            r = random.randint(0,l-1)
            n = remainingNumbers[r]
            boxes.append(n)
            del remainingNumbers[r]
        ## Assume success unless someone fails to find a number
        success = True
        for p in prisoners:
            numGuesses = 0
            found = False
            ## Set the first try to their number
            n = p
            while(numGuesses<maxGuesses):
                numGuesses += 1
                g = boxes[n]
                ## Set the new box num to try to what we just found
                ## If the guess is their number, then set found to True and break
                if (g==p):
                    found = True
                    break
                n = g
            if not found:
                success = False
                break
        attempts += 1
        if (success):
            successes += 1
    print("Successes:", successes, "/", attempts)
    
    
```
